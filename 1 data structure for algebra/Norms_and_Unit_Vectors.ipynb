```python
# 2.buat vector v dengan elemen (3,4)
v = np.array([3, 4])
v
```

```python
(3**2 + 4**2)**(1/2)
```

```python
# hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
# tampilkan unit vector dari vector v
v = np.array([3, 4])

norm = np.linalg.norm(v)
unit_v = v / norm

print("vektor asli:", v)
print("panjang vektor:", norm)
print("unit vektor", unit_v)
print("panjang unit vektor:", np.linalg.norm(unit_v))
```

```python
# hitung panjang unit vector dari vector v
norm = np.linalg.norm(v)
unit_v = v / norm
```
